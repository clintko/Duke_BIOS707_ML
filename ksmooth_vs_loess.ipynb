```r
library(tidyverse)
```

```r
set.seed(0)
x <- seq(1:10)
y <- x + rnorm(length(x), mean = 0, sd = 100)

options(repr.plot.height = 3, repr.plot.width = 3)
plot(x, y, pch = 20)
```

```r
fit_ks = ksmooth(x, y, kernel = c("box"), bandwidth = 2, n.points = 10)

plot(x, y, pch = 20)
points(fit_ks$x, fit_ks$y, col = "red", pch = 20)
points(x, y, pch = 20)
```

```r
fit_ks = ksmooth(x, y, kernel = c("box"), bandwidth = 2, n.points = 10)

plot(x, y, pch = 20)
points(fit_ks$x, fit_ks$y, col = "red", pch = 20)
points(x, y, pch = 20)

a = x[2]
b = mean(y[1:3])
points(a, b, col = "blue", pch = 20)
```

```r
fit_ls = loess(y ~ x,  degree = 0, span = 0.4)
y_ls = predict(fit_ls, data.frame(x = x))

plot(x, y, pch = 20)
points(fit_ks$x, fit_ks$y, pch = 20, col = "red")
points(x, y_ls, pch = 20, col = "blue")
```

```r
x1 = x[1:5]
y1 = y[1:5]
fit = lm(y1 ~ 1)

y1_fit = predict(fit, data.frame(x = 1))


plot(x, y, pch = 20)
points(fit_ks$x, fit_ks$y, pch = 20, col = "red")
points(x, y_ls, pch = 20, col = "blue")
points(1, y1_fit, pch = 20, col = "green")
```
